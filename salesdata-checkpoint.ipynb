```python
import pandas as pd
import matplotlib.pyplot as  plt
```

```python
df = pd .read_excel("sales_data.xlsx")
df
```

```python
import os 
print(os.getcwd())
```

```python
#create Total Sales column 
df.columns = df.columns.str.strip()
df["Total_Sales"] = df["Quantity"] * df["Price"]

df.groupby("Category")["Total_Sales"].sum().plot(kind="bar")
plt.title("Category Wise Sales")
plt.show()
```

```python
df.groupby("Region")["Total_Sales"].sum().plot(kind="bar")
plt.title("Region Wise Sales")
plt.show()
```

```python
df.groupby("Payment")["Total_Sales"].sum().plot(kind="pie",autopct="%1.1f%%")
plt.title("Payment Analysis")
plt.ylabel("")
plt.show()
```

```python
# Save final file for Power Bi
df.to_excel("clean_sales_data.xlsx",index=False)
print("file saved successfully")
```
